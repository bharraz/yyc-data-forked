```julia
using PyPlot
```

```julia
PyPlot.matplotlib["rcParams"][:update](Dict("font.size" => 40,
                                            "font.weight" => "bold"))
PyPlot.matplotlib[:rc]("xtick", labelsize=20)
PyPlot.matplotlib[:rc]("ytick", labelsize=20)
```

```julia
const nbar_axial = [15, 5.8, 2.7, 0.99, 0.425, 0.017]
const nbar_radial = [4, 1, 0.57, 0.45, 0.062, 0.028]
const pgrd = [0.002, 0.014, 0.037, 0.126, 0.585, 0.992];
```

```julia
figure(figsize=(20.0, 3))
plot(0:5, pgrd, "o-", lw=3, markersize=15)
gca()[:set_xticklabels]([])
grid()
ylim([0, 1.05])
xlim([-0.1, 5.1])
ylabel("Ground state\nprobability", size=25)
savefig("ground_state.png", bbox_inches="tight", transparent=true)
savefig("ground_state.svg", bbox_inches="tight", transparent=true)
```

```julia
figure(figsize=(20.0, 3))
plot(0:5, nbar_axial, "o-", lw=3, markersize=15, label="Axial")
plot(0:5, nbar_radial, "o-", lw=3, markersize=15, label="Radial")
gca()[:set_xticklabels]([])
grid()
gca()[:set_yscale]("log")
xlim([-0.1, 5.1])
ylim([0.011, 100])
ylabel("nbar")
legend(fontsize=30, ncol=2)
savefig("nbar.png", bbox_inches="tight", transparent=true)
savefig("nbar.svg", bbox_inches="tight", transparent=true)
```
