```julia
using NaCsData
using PyPlot
```

```julia
params, ratios, uncs = NaCsData.calc_survival("data/data_20170222_002525.csv");
```

```julia
TsOPProgress = linspace(0, 400e-6, 17)
offset1 = length(TsOPProgress)
TsOPHeating = linspace(0, 200e-3, 17)
offset2 = offset1 + length(TsOPHeating)
offset3 = offset2 + length(TsOPHeating)
TsD1Depump = linspace(0, 100e-3, 21)
```

# OP time

```julia
errorbar(TsOPProgress * 1e6, ratios[1:offset1, 2], uncs[1:offset1, 2])
xlabel("TOP (\$\\mu s\$)")
grid()
ylim([0, ylim()[2]])
title("OP time")
savefig("imgs/20170222/op_time.png", bbox_inches="tight", transparent=true)
savefig("imgs/20170222/op_time.svg", bbox_inches="tight", transparent=true)
```

# OP heating

```julia
errorbar(TsOPHeating * 1e3, ratios[(offset1 + 1):offset2, 2], uncs[(offset1 + 1):offset2, 2], label="0.02")
errorbar(TsOPHeating * 1e3, ratios[(offset2 + 1):offset3, 2], uncs[(offset2 + 1):offset3, 2], label="0.01")
xlabel("TOP (\$ms\$)")
legend()
grid()
ylim([0, ylim()[2]])
savefig("imgs/20170222/op_heating.png", bbox_inches="tight", transparent=true)
savefig("imgs/20170222/op_heating.svg", bbox_inches="tight", transparent=true)
```

# OP depumping

```julia
errorbar(TsD1Depump * 1e3, ratios[(offset3 + 1):end, 2], uncs[(offset3 + 1):end, 2])
xlabel("TOP (\$ms\$)")
grid()
title("Depumping with D1 F2")
ylim([0, ylim()[2]])
savefig("imgs/20170222/op_depumping.png", bbox_inches="tight", transparent=true)
savefig("imgs/20170222/op_depumping.svg", bbox_inches="tight", transparent=true)
```
