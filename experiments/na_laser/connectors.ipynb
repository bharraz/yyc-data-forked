## Thorlab Temperature Controller 9 Pin output

IIRC, Thorlabs and TimeBase use different conventions for TEC polarity

* 1: **Status LED+**
* 2: **Thermistor+**
* 3: **Thermistor-, GND**
* 4: _**TEC+**_
* 5: _**TEC-, Status LED-**_
* 6: **NC**
* 7: **AD 590/592-, LM 135/335+**
* 8: **LM 135/335-, GND**
* 9: **AD 590/592+, LM 135/335+**


## Thorlab Current Controller output

For connecting Timebase laser, we short the interlock (between 1 and 5) and use the Cathode ground mode (connect the cathode to Laser Diode GND and the anode to Laser Diode Anode).

* 1: **Interlock, Status ON/OFF**
* 2: **Photodiode**
* 3: **Laser Diode GND**
* 4: **Photodiode**
* 5: **GND (for 1)**
* 6: **Voltage Measurement Laser Diode Cathode**
* 7: **Laser Diode Cathode**
* 8: **Laser Diode Anode**
* 9: **Voltage Measurement Laser Diode Anode**
