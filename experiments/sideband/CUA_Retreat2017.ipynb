```julia
push!(LOAD_PATH, joinpath(pwd(), "../../lib"))

const data_dir = joinpath(pwd(), "data")
const img_dir = joinpath(pwd(), "imgs")
dataf(name, ext=".csv") = joinpath(data_dir, "$name$ext")
imgf(name, ext=".png") = joinpath(img_dir, "$name$ext")

using NaCsData
using PyPlot

PyPlot.matplotlib["rcParams"][:update](Dict("font.size" => 15,
                                            "font.weight" => "bold"))
PyPlot.matplotlib[:rc]("xtick", labelsize=20)
PyPlot.matplotlib[:rc]("ytick", labelsize=20)
```

```julia
function load_data(name)
    params, ratio, uncs = NaCsData.calc_survival(dataf(name))
    return params, ratio[:, 2], uncs[:, 2]
end
function save_image(name)
    savefig(imgf(name, ".png"), bbox_inches="tight", transparent=true)
    savefig(imgf(name, ".svg"), bbox_inches="tight", transparent=true)
end
```

```julia
figure(figsize=(10.0, 3.0))
axial = load_data("axial-cua_retreat2017")
errorbar(axial[1][1:50] ./ 1e6 .+ 18.48, axial[2][1:50], axial[3][1:50], fmt="ko-")
text(-0.1, 0.35, "Heating\n1", color="red", size=20, weight="bold",
     multialignment="center")
annotate("Carrier", xy=(0.0, 0.35), xytext=(0.01, 0.43),
         arrowprops=Dict("facecolor"=>"black", "edgecolor"=>"none",
                         "width"=>5, "headwidth"=>10, "headlength"=>10),
         color="black", size=20, weight="bold")
annotate("Cooling 1", xy=(0.065, 0.3), xytext=(0.11, 0.37),
         arrowprops=Dict("facecolor"=>"green", "edgecolor"=>"none",
                         "width"=>5, "headwidth"=>10, "headlength"=>10),
         color="green", size=20, weight="bold")
annotate("Cooling 2", xy=(0.13, 0.28), xytext=(0.18, 0.30),
         arrowprops=Dict("facecolor"=>"blue", "edgecolor"=>"none",
                         "width"=>5, "headwidth"=>10, "headlength"=>10),
         color="blue", size=20, weight="bold")
grid()
ylim([0, 0.5])
xlim([-0.1, 0.45])
xlabel("Detuning (MHz)")
ylabel("Survival")
title("Axial sideband (no RSC)")
save_image("axial-cua_retreat2017")
```
