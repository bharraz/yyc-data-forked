# DDS Noise

## Test before closing the box

All channels output $11.8$ to $12$dBm at $100$MHz with amplitude $1$.

Noise floors are not higher than $-45$dBm.

Second order harmonics are around $-30$ to $-25$dBm. Third orders are too small to see.
